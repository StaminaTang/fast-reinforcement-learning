## MarkovDecisionProcessDataBunch
The first peice to the RL puzzle is finding an easy way to handle environments. Our goal is for it to be as easy to use as the `TabularDataBunch` and `ImageDatabunch`. 

This will be accomplished by getting all the env names and testing the DataBunch runs.

```python
from fast_rl.core.Envs import Envs
from fast_rl.core.MarkovDecisionProcess import MarkovDecisionProcessDataBunch

envs = Envs.get_all_latest_envs()
for env in envs:
    print(f'Testing {env}')
    env_databunch = MarkovDecisionProcessDataBunch.from_env(env, max_steps=50, num_workers=0)
    if env_databunch is None: continue

    epochs = 1  # Also known as episodes

    for epoch in range(epochs):
        for element in env_databunch.train_dl:
            env_databunch.train_ds.actions = env_databunch.train_ds.get_random_action()

        for element in env_databunch.valid_dl:
            env_databunch.valid_ds.actions = env_databunch.valid_ds.get_random_action()
```
